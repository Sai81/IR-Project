```python
import os
notebook_path = os.path.abspath("IR_Project")
with open(os.path.join(os.path.dirname(notebook_path),"butterfly.json")) as json_file:
    data = json.load(json_file)

print(type(data))
print(data)
```

```python
filename_to_caption = {}
filename_to_obj_area = {}

for file_obj in data:
  filename = "/content"
  filename += file_obj['filename'][2:]
  img = PIL.Image.open(filename)
  wid, hgt = img.size
  image_area = wid * hgt
  objects = file_obj['objects']
  object_list_str = ""
  obj_with_area = {}

  for indi_object in objects:
    object_list_str += indi_object['name'] + " "
    area_of_box  = indi_object['relative_coordinates']['width'] * indi_object['relative_coordinates']['height']
    print("Area of ", indi_object['name'], " = ", area_of_box, "\t Image Area : ", image_area)
    obj_with_area[indi_object['name']] = area_of_box / image_area

  filename_to_caption[filename] = object_list_str
  filename_to_obj_area[filename] = obj_with_area
```
